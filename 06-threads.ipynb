```python
import _thread
import time

def thread_func(threadName, delay):
    count = 0
    while count < 5:
        time.sleep(delay)
        count += 1
        print(threadName + "  " + time.ctime(time.time()))
    
thread_func("1", 1)
thread_func("2", 2)
print("W" + "  " + time.ctime(time.time()))
```

```python
t1=_thread.start_new_thread(thread_func, ("1", 1))
t2=_thread.start_new_thread(thread_func, ("2", 2))
#thread_func("2", 2)
print("W" + "  " + time.ctime(time.time()))
```

1. Zrównoleglij wykonanie zaimplementowanej metody

```python
%%timeit
import sys
def find_min(l, lo, hi):
    minF = sys.maxsize
    if hi - lo < 1000:
        for k in range(lo, hi):
            if l[k] < minF:
                minF=l[k]
        return minF
    min1 = find_min(l, lo, int(lo + (hi - lo)/2))
    min2 = find_min(l, int(lo + (hi - lo)/2), hi)
    if min1 < min2:
        return min1
    else:
        return min2
print(find_min(mylist, 0, 100000))
```
