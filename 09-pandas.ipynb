## Series

Seria to struktura jednowymiarowa. Każdy element może mieć przypisaną wartość indeksu. Można je tworzyć z innych struktur danych.

```python
pd.Series(np.random.randn(5))
```

```python
pd.Series(np.random.randn(5), index=['a', 'b', 'c', 'd', 'e'])
```

```python
pd.Series({'a' : 0., 'b' : 1., 'c' : 2.})
```

```python
s = pd.Series(np.random.randn(5), index=list('abcde'))
s
```

```python
# pobieranie danych

print('s[1] = \n{}'.format(s[1]))
print('s[2:] = \n{}'.format(s[2:]))
print('s[1:-2] = \n{}'.format(s[1:-2]))
```

```python
# pobieranie danych z wykorzystaniem indeksu

print('s["b"] = \n{}'.format(s['b']))
print('s["c":] = \n{}'.format(s['c':]))
print('s["b":"c"] = \n{}'.format(s['b':'c']))
```

```python
# na seriach można wykonywać operacje

print('s*5 = \n{}'.format(s*5))
print('s**3 = \n{}'.format(s**3))
print('s*s = \n{}'.format(s*s))
print('s+s = \n{}'.format(s+s))
```

```python
# oraz je rysować
%matplotlib inline

s.plot()
```

### Zadanie

Utwórz DataFrame `samochody` z tysiącem losowych danych:
- `przebieg` z przedziału [0, 200 000]
- `spalanie` z przedziału [2, 20]

Dodaj kolumnę `marka`:
- jeżeli samochód ma spalanie [0, 8] marka to VW
- jeżeli samochód ma spalanie [8, 12] marka to Ford
- jeżeli samochód ma spalanie 13 i więcej, marka to UAZ
    
Dodaj kolumnę `pochodzenie`:
- jeżeli przebieg poniżej 100 km, pochodzenie `new`
- jeżeli przebieg powyżej 100 km, pochodzenie `PL`
- jeżeli przebieg powyżej 100 000 km, pochodzenie `DE`
- zapisz dane do pliku

Przeanalizuj dane statystycznie:
- pogrupuj dane po marce i po pochodzeniu (wskazówka: użyj `groupby(['marka', 'pochodzenie'])` i wskaż wybraną funkcję statystyczną (np. średnią)
- narysuj wykres przeanalizowanych danych (wskazówka: użyj wykresu typu `bar`, wskaż drugą oś y `secondary_y=['przebieg']`)
- sprawdź liczność grup i narysuj wykres
- wykonaj analizę statystyczną

```python
df3 = pd.DataFrame({ 'Przebieg' : np.random.randint(0, 200000, 1000), 'Spalanie' : np.random.randint(2, 20, 1000)})
df3.head(10)
```

```python
df3 = pd.DataFrame({ 'Przebieg' : np.random.randint(0, 200000, 1000), 'Spalanie' : abs(np.random.randint(2, 20, 1000)})
df3.head(10)
```

```python
samochody = pd.DataFrame({ 'przebieg' : np.random.randint(0, 200000, 1000), 'spalanie' : abs(np.random.rand(1000))*18+2})
samochody.head(10)
```

```python
samochody.loc[samochody.spalanie <= 8.0, 'marka'] = 'VW'
samochody.loc[samochody.spalanie > 8.0, 'marka'] = 'Ford'
samochody.loc[samochody.spalanie > 12.0, 'marka'] = 'UAZ'

samochody.loc[(samochody.przebieg >= 0) & (samochody.przebieg <100), 'pochodzenie'] = 'new'
samochody.loc[samochody.przebieg >= 100, 'pochodzenie'] = 'PL'
samochody.loc[samochody.przebieg >= 100000, 'pochodzenie'] = 'DE'

samochody.head(10)
```

```python
m = samochody.groupby(['marka', 'pochodzenie']).mean()
m.plot.bar(secondary_y=['przebieg']);
```
